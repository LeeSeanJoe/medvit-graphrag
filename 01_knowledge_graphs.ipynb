# 01 — Knowledge Graph Construction

Builds four pathology subgraphs following the RadGraph schema (Jain et al., NeurIPS 2021).
Each graph covers: Pleural Effusion, Cardiomegaly, Edema, Lung Opacity.

**Schema:**
- `located_at` (Observation → Anatomy): spatial location of a finding
- `suggestive_of` (Observation → Observation): diagnostic inference
- `modify` (Observation/Anatomy → Observation/Anatomy): quantification or characterization

Every edge is tagged `spatial=True/False`. Spatial edges drive Condition D (spatially-weighted GraphRAG).

**Sources:** 19 peer-reviewed radiology references — StatPearls, RadioGraphics, PMC, Radiology Masterclass.
Full citation list in `references.md`.

**Output:** `$DRIVE_BASE/knowledge_graphs/pathology_graphs.json`

```python
# serialise and save
serialisable = {
    pathology: {
        'nodes': graph['nodes'],
        'edges': [[e[0], e[1], e[2], e[3]] for e in graph['edges']]
    }
    for pathology, graph in KNOWLEDGE_GRAPHS.items()
}

out_path = f'{DRIVE_BASE}/knowledge_graphs/pathology_graphs.json'
with open(out_path, 'w') as f:
    json.dump(serialisable, f, indent=2)

print('Knowledge graph summary:')
print(f'{"Pathology":<20} {"nodes":>6} {"edges":>6} {"spatial":>8} {"non-spatial":>12}')
print('-' * 56)
for pathology, graph in KNOWLEDGE_GRAPHS.items():
    n_spatial    = sum(1 for e in graph['edges'] if e[3])
    n_nonspatial = len(graph['edges']) - n_spatial
    print(f'{pathology:<20} {len(graph["nodes"]):>6} {len(graph["edges"]):>6} '
          f'{n_spatial:>8} {n_nonspatial:>12}')
print(f'\nSaved to {out_path}')
```

```python
# validation: check all edge endpoints exist as nodes
print('Validating edge integrity...')
all_valid = True
for pathology, graph in KNOWLEDGE_GRAPHS.items():
    for edge in graph['edges']:
        src, rel, tgt, spatial = edge
        if src not in graph['nodes']:
            print(f'  ERROR [{pathology}]: source node "{src}" not in nodes')
            all_valid = False
        if tgt not in graph['nodes']:
            print(f'  ERROR [{pathology}]: target node "{tgt}" not in nodes')
            all_valid = False
if all_valid:
    print('All edges valid.')

# check spatial edge coverage: every anatomy node should appear in at least one spatial edge
print('\nAnatomy node spatial coverage:')
for pathology, graph in KNOWLEDGE_GRAPHS.items():
    spatial_nodes = set()
    for e in graph['edges']:
        if e[3]:
            spatial_nodes.add(e[0])
            spatial_nodes.add(e[2])
    anatomy_nodes = {k for k, v in graph['nodes'].items() if v['type'] == 'Anatomy'}
    uncovered = anatomy_nodes - spatial_nodes
    if uncovered:
        print(f'  [{pathology}] uncovered anatomy nodes: {uncovered}')
    else:
        print(f'  [{pathology}] all anatomy nodes covered')
```
